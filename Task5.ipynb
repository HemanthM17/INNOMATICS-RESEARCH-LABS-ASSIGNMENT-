```python
def post_engagement(data):
    total = sum(data)
    print('Total Likes:', total)
    if total >= 1000:
        print('Post Status: Viral Post')
    else:
        print('Post Status: Normal Engagement')

likes = [120, 180, 220, 90]
post_engagement(likes)
```

```python
def medicine_status(quantity):
    print('Medicine Stock:', quantity)
    if quantity < 10:
        print('Status: Low Stock Alert')
    else:
        print('Status: Stock Sufficient')

stock_value = 15
medicine_status(stock_value)
```

```python
def rainfall_check(data, required):
    avg = sum(data) / len(data)
    print('Average Rainfall:', int(avg))
    if avg >= required:
        print('Rainfall Status: Adequate Rainfall')
    else:
        print('Rainfall Status: Inadequate Rainfall')

rain_data = [40, 55, 60, 50]
required_level = 65
rainfall_check(rain_data, required_level)
```

```python
def appointment_check(age):
    print('Patient Age:', age)
    if age >= 18:
        print('Eligibility Status: Eligible')
    else:
        print('Eligibility Status: Not Eligible')

age_value = 16
appointment_check(age_value)
```

```python
def duplicate_accounts(users):
    if len(set(users)) != len(users):
        print('Duplicate Accounts Found: Yes')
    else:
        print('Duplicate Accounts Found: No')

user_list = ['alpha', 'beta', 'gamma', 'delta']
duplicate_accounts(user_list)
```

```python
def premium_filter(values):
    result = []
    for item in values:
        if item > 2000:
            result.append(item)
    print('Premium Crops:', result)

prices = [2100, 1800, 2600, 1400]
premium_filter(prices)
```

```python
def application_health(errors):
    print('Error Count:', errors)
    if errors == 0:
        print('System Status: Healthy')
    elif errors <= 5:
        print('System Status: Minor Issues')
    else:
        print('System Status: Critical Issues')

error_value = 3
application_health(error_value)
```

```python
def transaction_check(amount):
    limit = 50000
    print('Transaction Amount:', amount)
    if amount <= limit:
        print('Transaction Status: Approved')
    else:
        print('Transaction Status: Rejected')

amount_value = 45000
transaction_check(amount_value)
```

```python
def exam_check(records):
    total = len(records)
    attended = sum(records)
    percent = (attended / total) * 100
    print('Attendance Percentage:', percent)
    if percent >= 75:
        print('Exam Eligibility: Eligible')
    else:
        print('Exam Eligibility: Not Eligible')

attendance = [1, 0, 1, 0, 1]
exam_check(attendance)
```

```python
def electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        level = 'Low Usage'
    elif bill <= 1500:
        level = 'Moderate Usage'
    else:
        level = 'High Usage'

    return bill, level

units_used = 180
total, status = electricity_bill(units_used)
print('Units Consumed:', units_used)
print('Total Bill: ₹', total)
print('Usage Level:', status)
```
