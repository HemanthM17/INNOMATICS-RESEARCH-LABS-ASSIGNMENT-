Problem 1: Smart Parking Lot Management System

```python
def manage_parking():
    capacity = int(input('Enter parking capacity: '))
    logs = input('Enter vehicle logs (IN/OUT separated by space): ').split()
    current = 0
    peak = 0
    for entry in logs:
        if entry.upper() == 'IN' and current < capacity:
            current += 1
        elif entry.upper() == 'OUT' and current > 0:
            current -= 1
        peak = max(peak, current)
    print('Currently Parked Vehicles:', current)
    if current >= capacity:
        print('Parking Status: Full')
    else:
        print('Parking Status: Available')
    print('Peak Parking Usage:', peak)

manage_parking()
```

Problem 2: Online Food Delivery Time Estimator

```python
def estimate_delivery_time():
    distance = float(input('Enter distance in km: '))
    traffic = input('Enter traffic level (Low/Medium/High): ')
    weather = input('Enter weather condition (Clear/Rainy/Storm): ')
    total = distance * 5
    if traffic.lower() == 'high':
        total += 15
    elif traffic.lower() == 'medium':
        total += 10
    else:
        total += 5
    if weather.lower() == 'rainy':
        total += 10
    elif weather.lower() == 'storm':
        total += 20
    print('Estimated Delivery Time:', int(total), 'minutes')

estimate_delivery_time()
```

Problem 3: Movie Theatre Seat Occupancy Analyzer

```python
def analyze_seats():
    total_seats = int(input('Enter total seats: '))
    booked = list(map(int, input('Enter booked seat numbers separated by space: ').split()))
    booked_count = len(booked)
    percent = (booked_count / total_seats) * 100
    print('Occupancy:', int(percent), '%')
    if percent == 100:
        print('Show Status: Housefull')
    elif percent >= 70:
        print('Show Status: Almost Full')
    else:
        print('Show Status: Seats Available')

analyze_seats()
```

Problem 4: Cloud Server Load Classification System

```python
def classify_server_load():
    readings = list(map(int, input('Enter CPU readings separated by space: ').split()))
    avg = sum(readings) / len(readings)
    print('Average CPU Load:', int(avg), '%')
    if avg < 50:
        print('Server Status: Normal')
    elif avg <= 80:
        print('Server Status: Warning')
    else:
        print('Server Status: Critical')

classify_server_load()
```

Problem 5: Smart Classroom Resource Usage Monitor

```python
def monitor_resources():
    n = int(input('Enter number of resources: '))
    usage = {}
    for _ in range(n):
        name = input('Resource name: ')
        hours = float(input('Usage hours: '))
        usage[name] = hours
    overused = [r for r in usage if usage[r] > 8]
    if overused:
        print('Overused Resources:', ', '.join(overused))
        print('Energy Alert: Yes')
    else:
        print('Overused Resources: None')
        print('Energy Alert: No')

monitor_resources()
```

Problem 6: Online Event Registration Capacity Controller

```python
def manage_registrations():
    capacity = int(input('Enter event capacity: '))
    total = int(input('Enter total registrations: '))
    if total > capacity:
        confirmed = capacity
        waitlisted = total - capacity
        status = 'Closed'
    else:
        confirmed = total
        waitlisted = 0
        status = 'Open'
    print('Confirmed Registrations:', confirmed)
    print('Waitlisted Users:', waitlisted)
    print('Registration Status:', status)

manage_registrations()
```
