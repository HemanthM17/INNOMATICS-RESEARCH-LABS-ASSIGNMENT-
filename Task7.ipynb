Problem 1: Smart Light Controller

```python
class SmartLight:

    def __init__(self, location):
        self.location = location
        self.mode = "OFF"

    def toggle(self, value):
        if value.upper() == "ON":
            self.mode = "ON"
        elif value.upper() == "OFF":
            self.mode = "OFF"

    def status(self):
        print(self.location, "light status:", self.mode)


light_obj = SmartLight("Study-Room")
light_obj.toggle("ON")
light_obj.status()
```

Problem 2: Employee ID Card System

```python
class Employee:

    def __init__(self, emp_name, emp_code, section):
        self.emp_name = emp_name
        self.emp_code = emp_code
        self.section = section

    def display(self):
        print("Employee Details")
        print("Name:", self.emp_name)
        print("Code:", self.emp_code)
        print("Section:", self.section)


emp_obj = Employee("Arun", "EMP475", "HR")
emp_obj.display()
```

Problem 3: Mobile Contact Record

```python
class Contact:

    def __init__(self, username, mobile):
        self.username = username
        self.mobile = mobile

    def show(self):
        print("Saved Contact")
        print("User:", self.username)
        print("Mobile:", self.mobile)


contact_obj = Contact("Siddartha", 628124569)
contact_obj.show()
```

Problem 4: Product Price Tag Generator

```python
class Product:

    def __init__(self, product_name, amount):
        self.product_name = product_name
        self.amount = amount

    def generate_tag(self):
        print("Item:", self.product_name)
        print("Cost: ₹", self.amount)


product_obj = Product("Realme Tablet", 1899)
product_obj.generate_tag()
```

Problem 5: Movie Rating Display System

```python
class Movie:

    def __init__(self, movie_name, rating_value):
        self.movie_name = movie_name
        self.rating_value = rating_value

    def display_rating(self):
        print("Movie Name:", self.movie_name)
        print("User Rating:", self.rating_value, "/5")


movie_obj = Movie("Dil Bechara", 4.6)
movie_obj.display_rating()
```

Problem 6: Delivery Address Manager

```python
class DeliveryDetails:

    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def print_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


order1 = DeliveryDetails("Vinod Paul", "Nalgonda")
order1.print_details()
```
